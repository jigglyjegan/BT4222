## Bivariate Analysis (BA)

```python
# Perform a boxplot using 'flat_type' vs 'resale_price'
plt.figure(figsize = (16,8))
sns.boxplot(df_final['flat_type'], df_final['resale_price'], order = ['1 ROOM', '2 ROOM','3 ROOM','4 ROOM', '5 ROOM', 'EXECUTIVE', 'MULTI-GENERATION'])
```

```python
# Perform a boxplot using "flat_model" and "resale_price"
plt.figure(figsize = (16,8))
sns.boxplot(df_final['flat_model'], df_final['resale_price'])
plt.xticks(rotation = 45)
```

```python
plt.figure(figsize = (16,8))
sns.boxplot(df_final['storey_range'], df_final['resale_price'])
plt.xticks(rotation = 45)
```

```python
# Perform a boxplot using "town" and "resale_price"
plt.figure(figsize = (16,8))
sns.boxplot(df_final['town'], df_final['resale_price'])
plt.xticks(rotation = 45)
```

```python
# Plot floor_area_sqm vs resale_price
plt.figure(figsize = (16,8))
sns.scatterplot(df_final['floor_area_sqm'], df_final['resale_price'])
```

```python
# Plot remaining_lease vs resale_price
plt.figure(figsize = (16,8))
sns.scatterplot(df_final['remaining_lease'], df_final['resale_price'])
```

## Hypothesis: Does resale price increase with closer proximity to MRT?

```python
plt.figure(figsize = (16,8))
sns.scatterplot(x = df_final['dist to nearest mrt'], y = df_final['resale_price'], hue = df_final['flat_type'])
```
